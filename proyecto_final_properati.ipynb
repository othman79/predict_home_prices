#### El modelo secuencial es una pila lineal de capas. Se Crea un modelo secuencial llamando keras_model_sequential() función y luego a una serie de layerfunciones:

### Formas de entrada
#### El modelo necesita saber qué forma de entrada debe esperar. Por esta razón, la primera capa en un modelo secuencial (y solo la primera, porque las siguientes capas pueden hacer inferencias de forma automática) necesita recibir información sobre su forma de entrada.
#### Como se ilustra, esto se hace pasando un input_shape argumento a la primera capa. Esta es una lista de números enteros o NULL entradas, donde NULL indica que se puede esperar cualquier número entero positivo. En input_shape, la dimensión del lote no está incluida.


### compilación
#### Antes de entrenar nuestro modelo, se configuro el proceso de aprendizaje, que se realiza a través de la compile() función. El cual recibe tres argumentos:
    Un optimizador. Este podría ser el identificador de cadena de un optimizador existente (por ejemplo, como "rmsprop" o "adagrad") o una llamada a una función de optimizador (por ejemplo  optimizer_adam()).

    Una función de pérdida. Este es el objetivo que el modelo intentará minimizar. Puede ser el identificador de cadena de una función de pérdida existente (p. Ej., "Categórica_cruzada entropía" o "mse") o una llamada a una función de pérdida (p loss_mean_squared_error(). 
    
    Una lista de métricas. Para cualquier problema de regresión, querrá establecerlo en metrics = c('mean_squared_error').

#### El término ensamblador significa grupo. Los métodos tipo ensamblador están formados de un grupo de modelos predictivos que permiten alcanzar una mejor precisión y estabilidad del modelo. Estos proveen una mejora significativa a los modelos de árboles de decisión.
### ¿Por qué surgen los ensambladores de árboles?
#### Así como todos los modelos, un árbol de decisión también sufre de los problemas de sesgo y varianza. Es decir, ‘cuánto en promedio son los valores predecidos diferentes de los valores reales’ (sesgo) y ‘cuan diferentes serán las predicciones de un modelo en un mismo punto si muestras diferentes se tomaran de la misma población’ (varianza).
#### Al construir un árbol pequeño se obtendrá un modelo con baja varianza y alto sesgo. Normalmente, al incrementar la complejidad del modelo, se verá una reducción en el error de predicción debido a un sesgo más bajo en el modelo. En un punto el modelo será muy complejo y se producirá un sobre-ajuste del modelo el cual empezará a sufrir de varianza alta.
#### El modelo óptimo debería mantener un balance entre estos dos tipos de errores. A esto se le conoce como “trade-off” (equilibrio) entre errores de sesgo y varianza. El uso de ensambladores es una forma de aplicar este “trade-off”.

    •	Random Forest se considera como la “panacea” en todos los problemas de ciencia de datos.
    •	Útil para regresión y clasificación.
    •	Un grupo de modelos “débiles”, se combinan en un modelo robusto.
    •	Sirve como una técnica para reducción de la dimensionalidad.
    •	Se generan múltiples árboles (a diferencia de CART).
    •	Para regresión, se toma el promedio de las salidas (predicciones) de todos los árboles.

# Capa densa de Keras explicada.

## <div style='text-align: lift;'> ¿Qué es una capa densa en una red neuronal? </div>

### <div style='text-align: justify;'> La capa densa es una capa de red neuronal que está conectada profundamente, lo que significa que cada neurona en la capa densa recibe información de todas las neuronas de su capa anterior. Se encuentra que la capa densa es la capa más utilizada en los modelos. </div>

### <div style='text-align: justify;'> En segundo plano, la capa densa realiza una multiplicación matriz-vector. Los valores utilizados en la matriz son en realidad parámetros que se pueden entrenar y actualizar con la ayuda de la retro propagación. </div>

### <div style='text-align: justify;'> La salida generada por la capa densa es un vector dimensional 'm'. Por lo tanto, la capa densa se usa básicamente para cambiar las dimensiones del vector. Las capas densas también aplican operaciones como rotación, escalado, traslación en el vector. La clase Dense de Keras es una implementación del bloque de construcción de red neuronal más simple: la capa completamente conectada </div>

## <div style='text-align: lift;'> ¿Qué es el error cuadrático medio RMSE?  </div>
### <div style='text-align: justify;'> El error cuadrático medio (RMSE) mide la cantidad de error que hay entre dos conjuntos de datos. En otras palabras, compara un valor predicho y un valor observado o conocido.  </div>

## <div style='text-align: lift;'> ¿Qué es Métricas de regresión en aprendizaje automático: R al cuadrado (R²)?</div>


### <div style='text-align: justify;'>  Podemos usar para evaluar un modelo y está estrechamente relacionada con la MSE, pero tiene la ventaja de estar libre de escala. Se calcula el coeficiente de determinación, R², para cuantificar el rendimiento de su modelo. El coeficiente de determinación de un modelo es una estadística útil en el análisis de regresión, ya que a menudo describe qué tan "bueno" es ese modelo para hacer predicciones.</div>


## <div style='text-align: lift;'>¿Qué es el Error absoluto medio MAE?</div>

### <div style='text-align: justify;'> El MAE, es el error que se calcula como un promedio de diferencias absolutas entre los valores objetivo y las predicciones. El MAE es una puntuación lineal, lo que significa que todas las diferencias individuales se ponderan por igual en el promedio. Por ejemplo, la diferencia entre 10 y 0 será el doble de la diferencia entre 5 y 0. Sin embargo, lo mismo no es cierto para RMSE.</div>


## <div style='text-align: lift;'>En conclusión, se tomara las 3 metricas RMSE, R2 y el MAE</div>

    * R² es la proporción entre lo bueno que es nuestro modelo y lo bueno que es el modelo medio ingenuo.
    * el MAE es más robusto (menos sensible a los valores atípicos) que el MSE, pero esto no significa que siempre sea mejor usar el MAE.
